```
from frimcla.index_features import generateFeatures
from frimcla.StatisticalComparison import statisticalComparison
from frimcla.train import train
from frimcla.prediction import prediction
```

```
datasetPath = "./malaria-train"
outputPath = "./output"
batchSize = 32 
verbose = False
measure = "accuracy" #You can use accuracy, f1, auroc, precision or recall 
trainingSize = 1
nSteps=10
featureExtractors = [["lab888"]]
modelClassifiers = "GradientBoost"
```

```
generateFeatures(outputPath, batchSize, datasetPath, featureExtractors, verbose)
```

We create the training file and train the model.

```
f = open("output/malaria-train/ConfModel.json", "w")
f.write("{\n\"featureExtractors\": [\n{\n\"model\": \"lab888\",\n\"params\": \"8,8,8\",\n\"classificationModels\": [\n\"GradientBoost\"\n],\n\"multiclass\": false\n}\n]\n}")
f.close()
```

```
train(outputPath, datasetPath, trainingSize)
```

Finally, we evaluate the model.

```
featExt = ["lab888"]
classi = "GradientBoost"
classes = []
predictions = []
from imutils import paths

for img in list(paths.list_files('malaria-test/Parasitized/')):
  try:
    p = prediction(featExt, classi, img, outputPath, datasetPath)
    predictions.append(p[0][0])
    classes.append('Parasitized')
  except:
    print(img)
    
for img in list(paths.list_files('malaria-test/Uninfected/')):
  try:
    p = prediction(featExt, classi, img, outputPath, datasetPath)
    predictions.append(p[0][0])
    classes.append('Uninfected')
  except:
    print(img)

```

```
from sklearn.metrics import accuracy_score
```

```
accuracy_score(classes,predictions)
```
